```python
from controller import (
    add_usuario, 
    excluir_usuario, 
    atualizar_usuario,
    exibir_informacoes_usuarios, 
    exibir_um_usuario
    )
from module import ler_arquivo


def adiconar_usuarios():
    novos_usuarios = []
    opcoes_confirmacao = ["sim", "s", "não", "n"]
    opcoes_afirmativa = opcoes_confirmacao[:2]
    opcoes_negativas =opcoes_confirmacao[2:]
    inserir = True
    while inserir:
        nome = 'Não Informado'
        telefone = 'Não Informado'
        endereco = 'Não Informado'
        cadastrar_nome = True
        while cadastrar_nome:
            questiona_cadastro_nome = input("Deseja informar o nome? (S/N)")
            if questiona_cadastro_nome.lower() in opcoes_afirmativa:
                cadastrar_nome = False
                nome = input("Digite o nome: ")
            elif questiona_cadastro_nome.lower() in opcoes_negativas:
                cadastrar_nome = False
            else:
                print("Opção inválida")
        
        cadastrar_telefone = True
        while cadastrar_telefone:
            questiona_cadastro_telefone = input("Deseja informar o telefone? (S/N)")
            if questiona_cadastro_telefone.lower() in opcoes_afirmativa:
                cadastrar_telefone = False
                telefone = input("Digite o telefone: ")
            elif questiona_cadastro_telefone.lower() in opcoes_negativas:
                cadastrar_telefone = False
            else:
                print("Opção inválida")
        
        cadastrar_endereço = True
        while cadastrar_endereço:
            questiona_cadastro_endereco = input("Deseja informar o endereço? (S/N)")
            if questiona_cadastro_endereco.lower() in opcoes_afirmativa:
                cadastrar_endereço = False
                endereco = input("Digite o endereço: ")
            elif questiona_cadastro_endereco.lower() in opcoes_negativas:
                cadastrar_endereço = False
            else:
                print("Opção inválida")
        novos_usuarios.append({
            "Nome": nome,
            "Telefone": telefone,
            "Endereço": endereco
            })
        
        novo_cadastro = input("Deseja cadastar outro usuário?(S?N): ")
        if novo_cadastro.lower() in opcoes_afirmativa:
            pass
        else:
            inserir = False
        
    for novo_usuario in novos_usuarios:
        try:
            add_usuario(**novo_usuario)
        except Exception as error:
            print(error)
    
    iniciarSistema()

def excluir():
    inserir = True
    while inserir:
        id = input("Informe o id: ")
        try:
            excluir_usuario(id)
            print("Usuário excluido")
           
        except Exception as error:
            print(error)
    iniciarSistema()
    

def alterar():
    usuarios = ler_arquivo()
    id = ''
    opcoes = ["1", "2", "3"]
    opcao_selecionada = ""
    dicionario_opcoes = {
        "1": "Nome",
        "2": "Telefone",
        "3": "Endereço"
    }
    informar_id = True
    novo_dado = ''
    while informar_id:
        id = input("Insira o ID do usuário")
        if id not in usuarios:
            print("Usuário não encontrado!")
            informar_id = True
        else:
            informar_id = False
    
    seleciona_opcao = True
    while seleciona_opcao:
        print("Qual informação deseja alterar?\n1 - Nome\n2 - Telefone\n3-endereço")
        opcao_selecionada = input("Digite a opção: ")
        if opcao_selecionada not in opcoes:
            print("Opção inválida")
        else:
            novo_dado = input(f"Insira o {dicionario_opcoes[opcao_selecionada]}: ")
            seleciona_opcao = False

    try:
        atualizar_usuario(id, {dicionario_opcoes[opcao_selecionada]: novo_dado})
        print("Usuário atualizado")
    except Exception as error:
        print(error)
    iniciarSistema()

def exibir_apenas_um_usuario():
    inserir_id = True
    id=''
    while inserir_id:
        id = input("Informe o id: ")
        try: 
            exibir_um_usuario(id)
            inserir_id = False
        except Exception as error:
            print(error)
    iniciarSistema()

def exibir_todos_usuarios_ativos():
    try:
        exibir_informacoes_usuarios()
    except Exception as error:
        print(error)
    iniciarSistema()


def sair():
    print("Até breve")
    return

        
def iniciarSistema():
    opcoes = ["1 - Inserir usuário",
                "2 - Excluir usuário",
                "3 - Atualizar usuário",
                "4 - Informações de um usuário",
                "5 - Informações de todos os usuários",
                "6 - Sair"]
    print("Boas vindas ao nosso sistema.\n")
    
    for opcao in opcoes:
        print(opcao, end="\n")

    dicionario_opcoes = {
        "1": adiconar_usuarios,
        "2": excluir,
        "3": alterar,
        "4": exibir_apenas_um_usuario,
        "5": exibir_todos_usuarios_ativos,
        "6": sair
    }
    opcao_selecionada = input("Digite o número da opção desejadata")

    dicionario_opcoes[opcao_selecionada]()


iniciarSistema()
```
